```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import os

path = kagglehub.dataset_download("rohiteng/amazon-sales-dataset")
csv_file_path = os.path.join(path, 'Amazon.csv')
df = pd.read_csv(csv_file_path)
display(df.head())

df['BaseRevenue'] = df['Quantity'] * df['UnitPrice']
avg_qty = df['Quantity'].mean()
df['HighVolumeOrder'] = np.where(df['Quantity'] > avg_qty, True, False)
financial_summary = df[['Quantity', 'UnitPrice', 'Discount', 'Tax', 'ShippingCost', 'TotalAmount']].describe()
display(financial_summary)

brand_performance = df.groupby(['Category', 'Brand']).agg(
Total_Sales=('Quantity', 'sum'),
Total_Revenue=('TotalAmount', 'sum'),
Avg_Discount=('Discount', 'mean')).sort_values(by='Total_Revenue', ascending=False).reset_index()
display(brand_performance.head(10))

avg_shipping_by_country = df.groupby('Country')['ShippingCost'].mean().reset_index()
display(avg_shipping_by_country)

plt.figure(figsize=(5, 4))
status_counts = df['OrderStatus'].value_counts()
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%',colors=['green', 'maroon', 'blue','orange','yellow'], startangle=140, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.title('Order Fulfillment Status Breakdown', fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.show()

plt.figure(figsize =(6,5))
sns.countplot(x='PaymentMethod', data=df, order=df['PaymentMethod'].value_counts().index, palette='muted', hue='PaymentMethod', legend=False)
plt.title('Most Preferred Payment Methods by Transaction Count', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Payment Method', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 7))
country_sales = df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='Country', y='TotalAmount', data=country_sales, palette='magma', hue='Country', legend=False)
plt.title('Total Revenue Generation by Country', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.figure(figsize=(6,4))
plt.tight_layout()
plt.show()
```
